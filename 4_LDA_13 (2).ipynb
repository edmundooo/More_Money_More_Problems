```python
from  sklearn.feature_extraction.text  import CountVectorizer

# create the document-term matrix with count vectorizer
cv = CountVectorizer()
matrix = cv.fit_transform(tweets.clean_text)
columns=cv.get_feature_names()
```

```python
matrix = matrix.transpose()
```

```python
len(columns)
```

```python
corpus = matutils.Sparse2Corpus(matrix)
```

```python
id2word = dict(zip(list(range(0, 423990)), columns))
```

## LDA

```python
lda.print_topics(num_topics=50)
```

```python
# Transform the docs from the word space to the topic space (like "transform" in sklearn)
lda_corpus = lda[corpus]
```

```python

# Store the documents' topic vectors in a list so we can take a peak# Store  
lda_docs = [doc for doc in lda_corpus]
```

```python
# add lda docs list as series to corresponding df (pivoted or normal and then pickle
```

```python
tweet_topics = pd.DataFrame()
```

```python
tweet_topics['tweets'] = tweets.permalink
```

```python
tweet_topics.reset_index(inplace=True)
```

```python
topics = []

for i in lda_docs:
    topics.append(dict(i))
```

```python
topics_df = pd.DataFrame(topics)
```

```python
topics_df.shape
```

```python
tweets_lda_50_topics = pd.concat([tweet_topics, topics_df], axis=1, sort=False)
```
